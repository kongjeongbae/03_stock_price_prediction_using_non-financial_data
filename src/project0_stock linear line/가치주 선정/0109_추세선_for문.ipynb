```python
import pandas as pd
import numpy as np
```

```python
a = pd.read_csv('data/stockcode.csv')
a.head()
```

```python
list_code = []
list_company = []
for i in range(len(a)):
    list_code.append(str(a['code'][i]).zfill(6))
    list_company.append(a['company'][i])
```

```python
list_code
```

```python
len(list_code)
```

```python
import scipy as sp
import scipy.stats
from scipy import stats
import matplotlib.pyplot as plt
```

```python
total_slope = []
total_intercept = []
for i in range(len(list_code)):
    b = pd.read_csv('data/kospi200/%s.KS.csv' % list_code[i])
    y = b['Close']
    X= np.array(range(len(y)))
    slope, intercept, r_value, p_value, std_err = stats.linregress(X,y)
    line = slope*X+intercept
    total_slope.append(slope)
    total_intercept.append(intercept)
    plt.plot(X,y,'o', X, line)
    plt.savefig("data/line/line%s.png" % list_code[i])
    plt.clf()
```

```python
total_slope 
```

```python
a["slope"] = total_slope
a["intercept"] = total_intercept
```

```python
a.head(10)
```

```python
high_line = a.sort_values("slope", ascending= False)
```
